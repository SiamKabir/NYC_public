```python
# Re-run this cell
import pandas as pd
```

```python
# Read in the data
schools = pd.read_csv("/content/drive/MyDrive/NYC Public school test result score/schools.csv")
```

```python
# Preview the data
schools.head()
```

```python
#Â Which schools are best for math?
BestMathSchools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
print(BestMathSchools)
```

```python
# Calculate total_SAT per school
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
print(schools)
```

```python
# Who are the top 10 performing schools?
top10schools = schools.sort_values("total_SAT", ascending=True)[["school_name", "total_SAT"]].head(10)
print(top10schools)
```

```python
# Which NYC borough has the highest standard deviation for total_SAT?
boroughs = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)
print(boroughs)
```

```python
# Filter for max std and make borough a column
largest_std_dev = boroughs[boroughs["std"] == boroughs["std"].max()]
print(largest_std_dev)
```

```python
# Rename the columns for clarity
largest_std_dev = largest_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
print(largest_std_dev)
```
